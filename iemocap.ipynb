```python
import pandas as pd
import os

```

```python
emo_files = os.listdir('filelists/iemocap')
df = pd.DataFrame()
for emo_file in emo_files:
    emo_db = pd.read_csv(os.path.join("filelists/iemocap",emo_file))
    emotion = emo_file.split(".")[0]
    emo_db["emotion"] = emotion
    df = pd.concat([df,emo_db],ignore_index=True)
```

```python
df["emotion"].value_counts().plot(kind='pie')
```

```python
df.to_csv("filelists/iemocap.csv",header=None)
```
